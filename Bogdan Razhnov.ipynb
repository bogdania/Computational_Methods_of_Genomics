```python
import numpy as np
from numpy.random import choice
from matplotlib import pyplot as plt
import random 
```

```python
#Задаем матрицу переходных вероятностей А
#Задаем матрицу эмиссий E
#Согласно условию дз
```

```python
A=np.array([[0.95,0.05],[0.1,0.9]])
E = np.array([[1/6,1/6,1/6,1/6,1/6,1/6],[1/10,1/10,1/10,1/10,1/10,1/2]])
```

```python
#Формируем вертора: Наблюдений - m, Состояний - k
```

```python
l=400#длина вектора реальных состояний
m=np.zeros((l,),dtype=int)
k=np.zeros((l,),dtype=int)
```

```python
#случайная последовательность при заданных параметрах E и A
def HMM(E, A, L):
    if 0 <= random.random() < (2 / 3):
        D = 0
    else:
        D = 1
    path = []
    s = []
    for i in range(l):
        X = random.random()
        K = random.random()
        if (D == 0) and (X < 0.05):
            D = 1
        elif (D == 1) and (X < 0.1):
            D = 0
        if D == 0:
            s.append(0)
            path.append(random.randint(1, 6))
        elif D == 1:
            s.append(1)
            if K < 0.5:
                path.append(6)
            else:
                path.append(random.randint(1, 5))
    return(path, s)


path, s = HMM(E, A, l)

#наблюдение:
print(path)

#состояние:
print(s)
```

```python
#Алгоритм Витерби
#Инициализация:
Vk = [[0]*l, [0]*l]
Pv = [[0]*(l- 1), [0]*(l - 1)]
Vk[0][0] = E[0][path[0] - 1]*(2 / 3)
Vk[1][0] = E[1][path[0] - 1]*(1 / 3)
#Рекурсия:
for i in range(1, l):
    Vk[0][i] = E[0][path[i] - 1]*max(Vk[0][i - 1]*A[0][0], Vk[1][i - 1]*A[1][0])
    Vk[1][i] = E[1][path[i] - 1]*max(Vk[1][i - 1]*A[1][1], Vk[0][i - 1]*A[0][1])
    if Vk[0][i - 1]*A[0][0] > Vk[1][i - 1]*A[1][1]:
        Pv[0][i - 1] = 0
    else:
        Pv[0][i - 1] = 1
    if Vk[1][i - 1]*A[1][0] > Vk[0][i - 1]*A[0][1]:
        Pv[1][i - 1] = 1
    else:
        Pv[1][i - 1] = 0

#Нахождение наиболее вероятной траектории:
piv = [0]*l
if Vk[0][l - 1] > Vk[1][l - 1]:
    piv[l - 1] = 0
else:
    piv[l - 1] = 1
for i in range(l - 1):
    piv[l - (i + 2)] = Pv[piv[l - (i + 1)]][l - (i + 2)]
    
sovp=0
for i in range(l):
    if s[i]==piv[i]:
        sovp=sovp+1
print('Количество совпадений:', sovp,'из',l,'бросков')
```

```python
def forward(A, pi, E, path):
    N = pi.shape[0]
    alpha = np.zeros((l, N))
    alpha[0, :] = piv * E[:,path[0]]
    for t in range(1, l):
        for j in range(N):
            for i in range(N):
                alpha[t, j] += alpha[t-1, i] * A[i, j] * E[j, path[t]]
    
    return np.sum(alpha[l-1,:])
```

```python
#Алоритм просмотра вперед
def forward(E, A, path, l):
    fk = [[0]*(l+1), [0]*(l+1)]
    fk[0][0] = fk[1][0] =1
    fk[0][1] = E[0][path[0]-1]*(2 / 3)
    fk[1][1] = E[1][path[0]-1]*(1 / 3)
    for i in range(2, l+1):
        fk[0][i] = E[0][path[i-1]-1]*(fk[0][i-1]*A[0][0] + fk[1][i-1]*A[1][0])
        fk[1][i] = E[1][path[i-1]-1]*(fk[1][i-1]*A[1][1] + fk[0][i-1]*A[0][1])
    return(fk)


fk = forward(E,A, path, l)
resf = fk[0][-1] + fk[1][-1]

print('Cостояние - честная кость:',fk[0],'Cостояние - нечестная кость:',fk[1],resf,sep='\n')
```

```python
#Алгоритм просмотра вперед
def forward2(transition,emission,L,observations):
    F_k = []#Вероятность наблюдения последовательности при требовании что путь равен k=len(transition)
    for i in range(0,len(transition)):
        F_k.append([0]*(L+1))
    F_k[0][0] = F_k[1][0] = 1
    F_k[0][1] = emission[0][observations[0]-1]*(2 / 3)
    F_k[1][1] = emission[1][observations[0]-1]*(1 / 3)
    i=2
    while i<=L:
        sum_1 = F_k[0][i-1]*transition[0][0] + F_k[1][i-1]*transition[1][0]
        sum_2 = F_k[1][i-1]*transition[1][1] + F_k[0][i-1]*transition[0][1]
        F_k[0][i] = emission[0][observations[i-1]-1]*sum_1
        F_k[1][i] = emission[1][observations[i-1]-1]*sum_2
        i += 1
    check = F_k[0][L] + F_k[1][L]
    return (F_k, check)
F_k,check = forward2(A,E,l,path)
#F_k[0] - честная кость F_k[1] -  кость
print('честная кость:', '\n',F_k[0], '\n','нечестная:', '\n', F_k[1], '\n','Проверка +:', '\n', check)
```

```python
#Алоритм просмотра назад
def backward2(transition,emission,L,observations):
    B_k = []# эквивалентно F_k, но вычисляется обратной рекурсией
    for i in range(0,len(transition)):
        B_k.append([0]*(L+1))
    B_k[0][L] = B_k[1][L] = 1
    i = L-1
    while i>=0:
        B_k[0][i] = (emission[0][observations[i]-1]*B_k[0][i+1]*transition[0][0] + emission[1][observations[i]-1]*B_k[1][i+1]*transition[0][1])
        B_k[1][i] = (emission[1][observations[i]-1]*B_k[1][i+1]*transition[1][1] + emission[0][observations[i]-1]*B_k[0][i+1]*transition[1][0])
        i -=1
    B_k[0][0] = B_k[0][0]*(2/3)
    B_k[1][0] = B_k[1][0]*(1/3)
    check = B_k[0][0] + B_k[1][0]
    return (B_k,check)
B_k, check =  backward2(A,E,l,path)
print(B_k[0], '\n', B_k[1], '\n', check)
```

```python
#Апостериорное декодирование
def aposter(F_k,B_k,L):
    post = [0]*(L+1)
    i=0
    while i<=L:
        P_x = F_k[0][L] + F_k[1][L]
        post[i] = F_k[0][i]*B_k[0][i]/P_x
        i += 1
    return (post)
post = aposter(F_k,B_k,l)
print(post)
```
